```python
#Ans 1 :

class vehicle :
    def __init__ (self, vehicle_name, maximum_speed, vehicle_avg):
        self.vehicle_name = vehicle_name
        self.maximum_speed = maximum_speed
        self.vehicle_avg = vehicle_avg
        
    def return_details(self):
        return self.vehicle_name, self.maximum_speed, self.vehicle_avg
        
Car1 = vehicle("Fortuner", 180, 10)
Car1.return_details()


print("================================================================================================\n")


#ANS 2 :

class Car(vehicle):
    def seating_cap(self, capacity):
        self.capacity = capacity
        
    def return_details(self):
        return self.capacity, self.vehicle_name
        
Car2 = Car("Fortuner", "5 Seating Capacity", 10)        
Car2.return_details()
```

```python
# ANS 4 :

''' GETTER is the method in python OOPS which is created to access the private intances of a class which is
    not possible for an external user or function to directly access the instance, it is also used to get 
    value of an instance variable.
    
    SYNTAX : def car_get()
    
    ------------------------------------------------------------------------------------------------------
    
    SETTER is the method in python OOPS which is created to set a specific value to the private intances of a 
    class.  
  
    SYNTAX : def car_get()
'''

class car :
    def __init__ (self, name, speed, price):
        self.name = name
        self.speed = speed
        self.price = price
        
    def car_set (self):
        self.name = "JAGUAR"
        self.speed = 230
        self.price = 9000000
        
    def car_get (self):
        return self.name, self.speed, self.price
    
C1 = car("Jaguar", 230, 9000000)
C1.car_get()
        
        
        
```

```python
#ANS 5 :

''' METHOD OVERRIDING is the when we pass different methods but of same name and signature
    having different formal arguments. A method called "def sum()" is created to add int, float, complex
    numbers , here arguments will be different for each SUM() method as different datatypes are there
    but to add them SUM() method is same. '''


class ADD1():
      
    def __init__(self, a,b):
        self.a = a
        self.b = b
        
    def add(self):
        return self.a + self.b
          
            
class ADD2 (ADD1):
      
    def __init__(self, x,y):
        self.x = x
        self.y = y
          
    def add(self):
        return self.x + self.y
        
        
add1 = ADD1(12,14)
add2 = ADD2(12.5, 13.5)
add1.add()
add2.add()
```
